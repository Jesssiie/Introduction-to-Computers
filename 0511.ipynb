```python
temp=float(input("請輸入您的體溫"))
if temp>=37.5:
  print("體溫過高!")
elif temp<37.4 and temp>=37:
  print("體溫略高!")
elif temp<36.9 and temp>=34:
  print("體溫正常!")
else:
  print("體溫異常，請重新量測!")
```

```python
money=int(input("請輸入購物金額"))
if money>=10000:
  if money>=100000:
    print("打八折",money*0.8,end="元\n")
  elif money>=50000:
    print("打八五折",money*0.85,end="元\n")
  elif money>=30000:
    print("打九折",money*0.9,end="元\n")
  else:
    print("打九五折",money*0.95,end="元\n")
else:
  print("沒打折",money,end="元\n")
```

```python
score=int(input("請輸入成績"))
if score>=90:
  print("A")
elif score<=89 and score>80:
  print("B")
elif score<=79 and score>70:
  print("C")
elif score<=69 and score>60:
  print("D")
else:
  print("E")
```

```python
month=int(input("請輸入月份"))
if month>=3 and month<=5:
  print(month,end="月是春天!\n")
elif month>=6 and month<=8:
  print(month,end="月是夏天!\n")
elif month>=9 and month<=11:
  print(month,end="月是秋天!\n")
elif month==12 or month==1 or month==2:
  print(month,end="月是冬天!\n")
else:
  print(month,end="月份不在範圍內!\n")
```

```python
income=int(input("請輸入今年收入淨額"))
if income>300000:
  if income>=300000 and income<600000:
    print("付稅金額:",income*0.06,end="元\n")
  if income>=600000 and income<1000000:
    print("付稅金額:",income*0.13,end="元\n")
  if income>=1000000 and income<2000000:
    print("付稅金額:",income*0.21,end="元\n")
  if income>=2000000:
    print("付稅金額:",income*0.3,end="元\n")
else:
  print("付稅金額:",income*0,end="元\n")
```
